```python
# Import the useful stuff
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
# Stuff for error calculation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Declaring the features and the target
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd', 'OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF', 'Fireplaces']
target = 'SalePrice'
```

```python
# Splitting the dataset into training and testing sets
X = df[features]
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("Length of the training dataset is: ",len(X_train))
print("Length of the testing dataset is: ",len(X_test))
```

```python
# Creating and training the Random Forest Regressor model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Making predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Calculating evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
# Defining a model for the full data 
rf_model_on_test_data = RandomForestRegressor(random_state=42)
rf_model_on_test_data.fit(X, y)
```

```python
# Predicting on the test data
test_data = pd.read_csv('test.csv')
test_x = test_data[features]
# test_preds = rf_model_on_test_data.predict(test_x)
test_data.info()
```

```python
# Imputing the data because of missing values in test.csv
imputer = SimpleImputer(strategy='mean')
test_x_imputed = imputer.fit_transform(test_x)
test_preds = rf_model_on_test_data.predict(test_x_imputed)
```

```python
# Saving the predictions in a csv format
output = pd.DataFrame({'Id': test_data.Id, 'SalePrice': test_preds})
output.to_csv('output_rf.csv', index=False)
```
